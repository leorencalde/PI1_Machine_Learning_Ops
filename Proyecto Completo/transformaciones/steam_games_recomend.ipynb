1. IMPORTACIÓN DE LIBRERIAS:

```python
import pandas as pd
```

2. CARGA DE DATASET:

```python
# Leer el archivo JSON y crear un DataFrame
df = pd.read_json('../json/steam_games.json', lines=True)
```

```python
df
```

```python
# Eliminar filas donde todas las columnas tienen valores None o NaN
df_cleaned = df.copy()
df_cleaned = df.dropna(how='all')
```

```python
df_cleaned
```

```python
# Convertir la columna 'price' a numérico, reemplazando las cadenas con NaN y luego con 0, usando .loc para evitar SettingWithCopyWarning
df_cleaned.loc[:, 'price'] = pd.to_numeric(df_cleaned['price'], errors='coerce').fillna(0)
```

```python
df_cleaned
```

```python
# Contar las filas que tienen None en la columna 'genres'
none_count = df_cleaned['genres'].isnull().sum()

print(f"Cantidad de filas con None en la columna 'genres': {none_count}")
```

```python
# Eliminar filas donde la columna 'genres' es None
df_cleaned = df_cleaned.dropna(subset=['genres'])
```

```python
df_cleaned
```

```python
# Utilizar explode para expandir la lista en filas separadas
df_cleaned_exploded = df_cleaned.explode('genres')
```

```python
df_cleaned_exploded
```

4. DATASET EN FORMATO PARQUET Y/O CSV:

```python
# Guardar el nuevo dataset en formato Parquet
df_cleaned_exploded.to_parquet('steam_games_recommend.parquet')
```

```python
# Guardar el nuevo dataset en un archivo CSV
df_cleaned_exploded.to_csv('steam_games_recomend.csv', index=False)
```
