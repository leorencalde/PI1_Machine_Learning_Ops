1. IMPORTACIÓN DE LIBRERIAS:

```python
import pandas as pd
```

2. CARGA DE DATASET:

```python
# Leer el archivo JSON y crear un DataFrame
df = pd.read_json('../json/steam_games.json', lines=True)
```

```python
df
```

```python
# Eliminar filas donde todas las columnas tienen valores None o NaN
df_cleaned = df.dropna(how='all')
```

```python
df_cleaned
```

```python
# Convertir la columna 'price' a numérico, reemplazando las cadenas con NaN y luego con 0, usando .loc para evitar SettingWithCopyWarning
df_cleaned.loc[:, 'price'] = pd.to_numeric(df_cleaned['price'], errors='coerce').fillna(0)
```

```python
df_cleaned
```

4. DATASET EN FORMATO PARQUET Y/O CSV:

```python
# Guardar el nuevo dataset en formato Parquet
df_cleaned.to_parquet('steam_games_transf.parquet')
```

```python
# Guardar el nuevo dataset en un archivo CSV
df.to_csv('steam_games_transf.csv', index=False)
```
